```python
import requests
import pandas as pd
from bs4 import BeautifulSoup
# Get the addresses for the childcares in a certain area 
def CCget(area):
    headers = {
            'User-agent':
            "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/99.0.4844.74 Safari/537.36"
        }
    response = requests.get("https://skoolopedia.com/app/childcare/area/{}".format(area), headers = headers).text
    soup = BeautifulSoup(response, 'html.parser')
    target = soup.find_all("p", class_ = "small")
    output = []
    for i in range(0,len(target)):
        output.append(target[i].get_text())
    output1 = pd.DataFrame(output, columns = ['address'])
    output1.to_csv("Childcare{}.csv".format(area))
    return(output1)
    
```

```python
# Just put the area name here and it will output the file with the addresses
CCget()
```
