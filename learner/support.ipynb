```python
#avg SSE is obtained from one validating against one fold, so we just take fold 1 for simplicity.
def count_rows(csv_file_path):
    df = pd.read_csv(csv_file_path)
    print(type(df))
    return df[df["k_fold"] == 1].count()["k_fold"]

def avg_price(csv_file_path):
    df = pd.read_csv(csv_file_path)
    col = df.resale_price
    return col.mean()

# Value is the previously obtained validation SSE value
# avg value of SSE in 1 row = avg value of SSE across 10 folds / number of validation rows in 1 fold
# avg prediction value = sqrt(avg value of SSE in 1 row)
# percentage estimate = avg prediction value / true avg value
def percentage_estimate(value, csv_file_path):
    count = count_rows(csv_file_path)
    avg = avg_price(csv_file_path)
    return math.sqrt(value / count) / avg
```

```python
import matplotlib.pyplot as plt

df = pd.read_csv("results-RFmaxdepth-fixed.csv")
plt.figure(figsize=(10, 10))
for depth in range(16,26):
    data = df.loc[df['max-depth'] == depth].drop(["max-depth"], axis = 1)
    x = data.loc[:,"trees"].tolist()
    y = data.drop(["trees"], axis = 1).values.tolist()

    plt.plot(x, y, label = f'Depth: {depth}')


plt.legend(loc = "upper right")
plt.show()
```
